```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')


df = pd.read_csv('datos.csv', delimiter=';')

ax = sns.boxplot(x="materia", y="nota", data=df)
plt.title("Distribuci√≥n de Notas")
plt.xlabel("Materias")
plt.ylabel("Notas")
plt.show()
```

```python


# Datos de ejemplo
tupla = tuple(df['aprobado'])
conjunto = set(tupla)
tupla_sin_duplicados = tuple(conjunto)

df1 = df.query(f'aprobado == "{tupla_sin_duplicados[0]}"')
valor1 = df1['aprobado'].value_counts().values[0]

df2 = df.query(f'aprobado == "{tupla_sin_duplicados[1]}"')
valor2 = df2['aprobado'].value_counts().values[0]

categorias = ['Aprobados','No Aprobados']
valores = [valor1, valor2]


startangle = 90
# Crear el gr√°fico de torta
plt.pie(valores, labels=categorias, startangle=startangle,autopct='%1.1f%%')
plt.title("Distribuci√≥n de Aprobados")
startangle = 20

# Mostrar el gr√°fico
plt.show()
```
